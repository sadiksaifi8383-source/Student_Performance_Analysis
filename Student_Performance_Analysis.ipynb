```python
import pandas as pd
import numpy as np
#Load data using Pandas.
df = pd.read_csv('/content/Student.csv')
df.head()
```

```python
print("\nsummary statistics.")
print(df.describe())

print("\nMaximum_Value")
df.max()

print("\ntop 10 performers.")
top = df.nlargest(10, "Result")
print(top)


print("\nsubject-wise highest scores.")
df.sort_values(by='Result', ascending=False).head(10)


print("\nAverage of overall subjects")
df["Average"] = (df["Maths"] + df["Physics"] + df["Chemistry"])/3
print(df)

print("\nAssign grades (A, B, C, D).")
def grade(Average):
    if Average >= 90:
        return "A"
    elif Average >= 80:
        return "B"
    elif Average >= 70:
        return "C"
    elif Average >= 50:
        return "D"
    else:
        return "F"

df["Grade"] = df["Average"].apply(grade)

print(df),


print("\nValue_Counts")
print(df.value_counts())

print("\nStudent Correlation")
df.select_dtypes(include=np.number).corr()

print("\nStudent Covariance")
df.select_dtypes(include=np.number).cov()

print("\nStudent Variance")
df.select_dtypes(include=np.number).var()

print("\nStudent Standard Deviation")
df.select_dtypes(include=np.number).std()
```
